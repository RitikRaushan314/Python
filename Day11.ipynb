```python
#access list items
numbers = [10,20,30,40,50,60,70,80]
print(numbers)
print(numbers[1])
print(numbers[2:6])
print(numbers[3:])
print(numbers[:5])
print(numbers[1::2])
print(numbers[::-1])
```

```python
#[::-1]  return reversed copy
#reverse() modifies the original list
list1 = [10, 20, 30, 40, 50]

# Using slicing
reversed_copy = list1[::-1]

# Using reverse()
list2 = list1.copy()
list2.reverse()

print("Original List      :", list1)
print("Using [::-1]       :", reversed_copy)
print("Using reverse()    :", list2)
```

```python
#shallow copy and deep copy
import copy

list1 = [10, 20, 30]

shallow = copy.copy(list1)
deep = copy.deepcopy(list1)

print("Original List :", list1)
print("Shallow Copy  :", shallow)
print("Deep Copy     :", deep)

print("\nAddresses:")
print("list1   :", id(list1))
print("shallow :", id(shallow))
print("deep    :", id(deep))
```

```python
#copy 
import copy

list1 = [10, 20, 30]

copy1 = list1.copy()          # Using copy()
copy2 = list1[:]              # Using slicing
copy3 = list(list1)           # Using list()
copy4 = copy.copy(list1)      # Shallow copy
copy5 = copy.deepcopy(list1)  # Deep copy

print("Original List :", list1)
print("copy()        :", copy1)
print("[:]           :", copy2)
print("list()        :", copy3)
print("copy.copy()   :", copy4)
print("deepcopy()    :", copy5)
```
