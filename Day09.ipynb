```python
#factorial using recursion
def fact(n):
    if n==1:
        return 1
    else:
        return n*fact(n-1)

print(fact(5))
```

```python
#sum of n numbers
def sum(n):
    if n==0:
        return 1
    else:
        return n+sum(n-1)

print(sum(5))
```

```python
#recursive list sum without sum() and loop 
l = [12, 5, 7, 10, 8]

def list_sum(lst):
    if len(lst) == 0:      # Base case
        return 0
    return lst[0] + list_sum(lst[1:])   # Recursive case

print(list_sum(l))
```

```python
#build your own map()
l = [1,2,3,4,5]

def add_one(x):
    return x + 1

def my_map(func, lst):
    result = []
    for i in lst:
        result.append(func(i))
    return result

print(my_map(add_one, l))
```

```python
#recursive string comparission  aaabbbccccdda into a3b3c4d2a1
s = "aaabbbccccdda"

def compress(s):
    if s == "":
        return ""

    i = 1
    while i < len(s) and s[i] == s[0]:
        i += 1

    return s[0] + str(i) + compress(s[i:])

print(compress(s))
```

```python
#recursive number to words 5072 into five zero seven two using recursive only
num = "5072"

words = ["zero", "one", "two", "three", "four",
         "five", "six", "seven", "eight", "nine"]

def number_to_words(n):
    if n == "":
        return

    print(words[int(n[0])], end=" ")
    number_to_words(n[1:])

number_to_words(num)
```

```python
#recurssive decimal to binary 
num = int(input("Enter a number: "))

def binary(num):
    if num == 0:
        return
    binary(num // 2)
    print(num % 2, end="")

binary(num)
```

```python
#employee salary processing salary in tuple increase salary by 10% using map, filter salaries > 60000 using filter(), find total salary using reduce()
salary = [(50000,),(60000,),(70000,)]
inc = list(map(lambda x : x[0]*1.10,salary))
print(inc)

fil = list(filter(lambda x : x[0]>60000,salary))
print(fil)

from functools import reduce
salary = [(50000,), (60000,), (70000,)]
total = reduce(lambda x, y: x + y[0], salary[1:], salary[0][0])
print(total)
```
