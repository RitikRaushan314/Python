```python
#abstarction
#crate an abstract class Vechile with an abstract method fuel_type(). 
#create PetrolCar and ElectricCar subclasses, each implementing fuel_type() differently.
from abc import ABC, abstractmethod


class Vehicle(ABC):
    @abstractmethod
    def fuel_type(self):
        pass

class PetrolCar(Vehicle):
    def fuel_type(self):
        print("Fuel Type: Petrol")

class ElectricCar(Vehicle):
    def fuel_type(self):
        print("Fuel Type: Electricity")

car1 = PetrolCar()
car2 = ElectricCar()

car1.fuel_type()
car2.fuel_type()
```

```python
#inheritance
#create a base class Person with name and age.
#create Teacher and Student subclasses,each adding one extra attribute (eg subject for teacher , roll_no for student)
#and overriding a common method role() to describe who they are.

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def role(self):
        print("I am a Person.")

class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject

    def role(self):
        print(f"I am a Teacher.\nName: {self.name}\nAge: {self.age}\nSubject: {self.subject}")

class Student(Person):
    def __init__(self, name, age, roll_no):
        super().__init__(name, age)
        self.roll_no = roll_no

    def role(self):
        print(f"I am a Student.\nName: {self.name}\nAge: {self.age}\nRoll No: {self.roll_no}")


t1 = Teacher("Ramesh", 40, "Mathematics")
s1 = Student("Ritik", 20, 101)
t1.role()
print()
s1.role()
```

```python
#polymorphism
#create Square and Trinagle classes,each with a draw() method that prints a different simple pattern (eg using *).
#put objects of both in a list and loop through, calling draw() on each.

class Square:
    def draw(self):
        print("Square")
        for i in range(1,4):
            print("* "*6)
class Triangle:
    def draw(self):
        print("Triangle")
        for i in range(1,7):
            print("* "*i)

s1 = Square()
t1 = Triangle()

s1.draw()
t1.draw()
```

```python
#1 smart library system
#  Design a system to issue and return books. Store book availability safely, 
#  allow members to borrow/return books and calculate late fees if applicable.

class Book:
    def __init__(self, title):
        self.title = title
        self.__available = True
        
    def is_available(self):
        return self.__available

    def issue(self):
        if self.__available:
            self.__available = False
            return True
        return False

    def return_book(self):
        self.__available = True

class Member:
    def __init__(self, name):
        self.name = name
        self.borrowed_book = None

    def borrow(self, book):
        if book.issue():
            self.borrowed_book = book
            print(f"{self.name} borrowed '{book.title}'")
        else:
            print(f"'{book.title}' is not available.")

    def return_book(self, days_late):
        if self.borrowed_book:
            fee = 0
            if days_late > 0:
                fee = days_late * 10      

            self.borrowed_book.return_book()
            print(f"{self.name} returned '{self.borrowed_book.title}'")
            print("Late Fee = Rs.", fee)
            self.borrowed_book = None
        else:
            print("No book to return.")


book1 = Book("Python Programming")
member1 = Member("Ritik")
member1.borrow(book1)
member1.return_book(3)      
```

```python
#2 Airline Ticket Reservation
#  create Economy,Business and First Class tickets. Calculate ticket fare based on class and baggage weight
#  print boarding details

class Ticket:
    def __init__(self, passenger, baggage):
        self.passenger = passenger
        self.baggage = baggage

    def calculate_fare(self):
        pass

    def boarding_details(self):
        print("\n------ Boarding Pass ------")
        print("Passenger :", self.passenger)
        print("Class     :", self.__class__.__name__)
        print("Baggage   :", self.baggage, "kg")
        print("Fare      : Rs.", self.calculate_fare())
        print("---------------------------")

class Economy(Ticket):
    def calculate_fare(self):
        return 3000 + (self.baggage * 50)

class Business(Ticket):
    def calculate_fare(self):
        return 6000 + (self.baggage * 40)

class FirstClass(Ticket):
    def calculate_fare(self):
        return 10000 + (self.baggage * 30)

t1 = Economy("Ritik", 15)
t2 = Business("Aman", 20)
t3 = FirstClass("Priya", 25)

tickets = [t1, t2, t3]

for ticket in tickets:
    ticket.boarding_details()
```

```python
#3 Restaurant Billing
#  Customers order food items. Generate the bill,apply GST and discount,and support Dinening and Takeaway orders differently.

class Order:
    def __init__(self, customer, amount):
        self.customer = customer
        self.amount = amount

    def generate_bill(self):
        pass

class Dining(Order):
    def generate_bill(self):
        gst = self.amount * 0.05      
        discount = self.amount * 0.10 
        total = self.amount + gst - discount

        print("\n----- Dining Bill -----")
        print("Customer :", self.customer)
        print("Food Bill: Rs.", self.amount)
        print("GST      : Rs.", gst)
        print("Discount : Rs.", discount)
        print("Total    : Rs.", total)

class Takeaway(Order):
    def generate_bill(self):
        gst = self.amount * 0.05      
        packing = 30                  
        total = self.amount + gst + packing

        print("\n----- Takeaway Bill -----")
        print("Customer :", self.customer)
        print("Food Bill: Rs.", self.amount)
        print("GST      : Rs.", gst)
        print("Packing  : Rs.", packing)
        print("Total    : Rs.", total)

o1 = Dining("Ritik", 1000)
o2 = Takeaway("Aman", 800)
orders = [o1, o2]
for order in orders:
    order.generate_bill()
```

```python
#4 Online Examination Portal
# Store student privately. conduct an exam, calculate marks, assign grades, and generate a report card.

class Student:
    def __init__(self, name, roll_no):
        self.__name = name          
        self.__roll_no = roll_no    
        self.marks = 0

    def conduct_exam(self, marks):
        self.marks = marks

    def calculate_grade(self):
        if self.marks >= 90:
            return "A+"
        elif self.marks >= 75:
            return "A"
        elif self.marks >= 60:
            return "B"
        elif self.marks >= 40:
            return "C"
        else:
            return "Fail"

    def report_card(self):
        print("Name     :", self.__name)
        print("Roll No. :", self.__roll_no)
        print("Marks    :", self.marks)
        print("Grade    :", self.calculate_grade())

s1 = Student("Ritik", 101)
s1.conduct_exam(88)
s1.report_card()
```
