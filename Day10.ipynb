```python
#count no. of letter, digit, space
str = "Ritik Raushan"
letter=0
digit=0
space=0
for i in str:
    global letter
    global digit
    global space
    if i.isalpha():
        letter+=1
    if i.isdigit():
        digit+=1
    if i.isspace():
        space+=1
print("Letter:",letter)
print("Digit:",digit)
print("Space:",space)
```
