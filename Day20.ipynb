```python
import time
import numpy as np
 
size = 1_000_000
py_list = list(range(size))
np_array = np.arange(size)
 
start = time.time()
sum([x * 2 for x in py_list])
print('Python list:', time.time() - start, 'seconds')
 
start = time.time()
np_array * 2
print('NumPy array:', time.time() - start, 'seconds')
```
