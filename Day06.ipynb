```python
#greet the student welcome Ritik for five student
def greet(*name):
    for i in name:
        print("Welcome",i)

greet("Ritik","Raushan","Aakash","Amar","Prakash")
```

```python
#calculator for a function +,-,*,/,//,%
def cal(a, b):
    print("Addition       :", a + b)
    print("Subtraction    :", a - b)
    print("Multiplication :", a * b)
    print("Division       :", a / b)
    print("Floor Division :", a // b)
    print("Modulus        :", a % b)

cal(4, 2)
```

```python
#largest no. out of two
def max(a,b):
    if(a>b):
        print(a,"is largest")
    else:
        print(b,"is largest")
max(10,20)
```

```python
#even or odd 
def evenodd(a):
    if (a%2==0):
        print("EVEN")
    else:
        print("ODD")
evenodd(15)
```

```python
#largest out of five
def large(*nums):
    m=0
    for i in nums:
        if (i>m):
            m=i
    print(m)
large(7,8,9,6,5)
```

```python
#write function using each types of parameter
print("Positional Parameter")
def sum(a,b):
    print(a+b)
sum(10,20)

print("\nKeyword Parameter")
def info(name,regno):
    print(f"Name: {name}\nRegno: {regno}")
info(name="Ritik",regno=12300915)

print("\nDefault Parameter")
def greet(name,message="Welcome"):
    print(name,message)
greet("Ritik")

print("\nVariable Argument")
def sum(*nums):
    sum=0
    for i in nums:
        sum+=i
    print(sum)
sum(1,2,3,4,5)

print("\nVariable Keyword Argument")
def details(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

details(name="Ritik", age=20, city="Delhi")
```
