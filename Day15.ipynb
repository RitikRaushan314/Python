```python
# Create a file and write introduction
with open("student.txt", "w") as f:
    f.write("My name is Ritik.\n")
    f.write("I am pursuing B.Tech in Computer Science.\n")
    f.write("I like Python Programming.\n")
```

```python
# Read and display file contents
with open("student.txt", "r") as f:
    print(f.read())
```

```python
# Append a new line
with open("student.txt", "a") as f:
    f.write("I want to become a Data Analyst.\n")
```

```python
# Count number of lines
with open("student.txt", "r") as f:
    lines = f.readlines()
print("\nTotal Lines:", len(lines))
```

```python
# Count number of words
with open("student.txt", "r") as f:
    text = f.read()

words = text.split()
print("Total Words:", len(words))
```

```python
# Search a word
word = input("\nEnter a word to search: ")

if word in text:
    print("Word Found")
else:
    print("Word Not Found")

```

```python
# Copy one file into another
with open("student.txt", "r") as source:
    data = source.read()

with open("copy.txt", "w") as destination:
    destination.write(data)
```

```python
# Convert file content to uppercase
print(data.upper())

with open("uppercase.txt", "w") as f:
    f.write(data.upper())

print("content saved in uppercase.txt")
```

```python
#OOPs
class Student:
    college = "LPU"

    def __init__(self,name,age):
        self.name=name
        self.age=age

    def display(self):
        print(self.name,self.age,self.college)

s1 = Student("Ritik",20)
s2 = Student("Raushan",21)

s1.display()
s2.display()

print(s1.__class__)
print(type(s1))
```
