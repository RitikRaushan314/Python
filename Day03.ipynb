```python
#print 1 to 10
for i in range(1,11):
    print(i)
```

```python
#Even Number
for i in range(2,11,2):
    print(i)
```

```python
#Odd Number
for i in range(1,11,2):
    print(i)
```

```python
#Reverse Number
for i in range(10,0,-1):
    print(i)
```

```python
#Multiplication Table
num=int(input("Multiplication of:"))
for i in range(1,11):
    print(num*i)
```

```python
#sum of first n number
n=int(input("Enter number to add upto"))
sum=0
for i in range(1,n+1):
    sum+=i
print(sum)
```

```python
#factorial
a=int(input("Enter number:"))
fact=1
if a>0:
    for i in range(1,a+1):
        fact*=i
    print(fact)
elif a==0:
    print("1")
else:
    print("Enter Valid Number")
```
