```python
import numpy as np
import pandas as pd
df = pd.read_csv("C:/Users/raush/ Python Batch/student_performance_project.csv")
print(df)
```

```python
print(df.info())
print(df.describe())
```

```python
print(df.isna().sum())
```

```python
df['Math'] = df['Math'].fillna(df['Math'].mean())
df['Science'] = df['Science'].fillna(df['Science'].mean())
df['English'] = df['English'].fillna(df['English'].mean())
print(df.isna().sum())
```

```python
print(df.duplicated().sum())
df = df.drop_duplicates()
print(df)
```

```python
print(df[df['City'] == 'Pune'])
```

```python
print(df[['Math', 'Science', 'English']].sum())
```

```python
df['Total'] = df[['Math', 'Science', 'English']].sum(axis=1)
print(df[['StudentID', 'Name', 'Total']])

df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1).round(2)
print(df[['StudentID', 'Name', 'Average']])

df['Rank'] = df['Total'].rank()
print(df[['Name', 'Total', 'Rank']].sort_values('Rank'))
```

```python
df['Result'] = df[['Math', 'Science', 'English']].apply(
    lambda x: 'Pass' if (x >= 35).all() else 'Fail',
    axis=1
)
print(df[['Name', 'Result']])
```

```python
df['Result'] = df[['Math', 'Science', 'English']].apply(
    lambda x: 'Pass' if (x >= 35).all() else 'Fail',
    axis=1
)
print(df[['Name', 'Result']].head())
```

```python
print(df.groupby('Gender')[['Math', 'Science', 'English']].sum())
print(df.groupby('City')[['Math', 'Science', 'English']].sum())
```

```python
report = df.groupby(['City', 'Gender']).agg(
    Students=('StudentID', 'count'),
    Avg_Math=('Math', 'mean'),
    Avg_Science=('Science', 'mean'),
    Avg_English=('English', 'mean'),
    Avg_Total=('Total', 'mean'),
    Highest_Total=('Total', 'max'),
    Lowest_Total=('Total', 'min')
).round(2)

print(report)
```

```python
top10 = df.nlargest(10, 'Total')
print(top10[['StudentID', 'Name', 'City', 'Gender', 'Total', 'Rank']])
```

```python
final_report = df[['StudentID','Name','Gender','City','Math','Science','English','Total','Average','Rank','Result']]
print(final_report)
```

```python
final_report.to_csv('Student_Report.csv', index=False)
print("Student report exported successfully.")
```
