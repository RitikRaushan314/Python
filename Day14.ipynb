```python
#31 design a mini model social-network data supporting friends,mutual friends and friends suggestions.
# Mini Social Network

network = {
    "ritik": {"raushan", "amar", "prakash"},
    "raushan": {"ritik", "amit"},
    "Aakash": {"ujjwal", "prakash"}
}

name = input("Enter your name: ")

if name in network:
    print("Friends:", network[name])

    # Mutual Friends
    print("\nMutual Friends:")
    for person in network:
        if person != name:
            mutual = network[name] & network[person]   # Intersection
            if mutual:
                print(f"{person} : {mutual}")

    # Friend Suggestions
    print("\nFriend Suggestions:")
    suggestion = set()

    for friend in network[name]:
        if friend in network:
            suggestion |= network[friend]      # Union of friends

    suggestion -= network[name]               # Remove existing friends
    suggestion.discard(name)                  # Remove self

    print(suggestion)

else:
    print("User not found.")
```
