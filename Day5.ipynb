```python
#Hospital Rooms
buildings = int(input("Enter no. of Buildings:"))
floors = int(input("Enter no. of Floors:"))
rooms = int(input("Enter no. of Rooms:"))
visited_rooms=0

for i in range(1,buildings+1):
    for j in range(1,floors+1):
        for k in range(1,rooms+1):
            visited_rooms+=1
print("No. of Rooms Visited:",visited_rooms)
```
