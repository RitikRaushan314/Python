```python
#11 Even sum combination
for i in range(0,2):
    for j in range(0,2):
        for k in range(0,2):
            if ((i+j+k)%2==0):
                print(i,j,k)
```

```python
#15. Multiplication table
def table(a,b):
    for i in range(a,b+1):
        print("Table:",i)
        for j in range(1,11):
            print(i*j)
table(1,3)
```
