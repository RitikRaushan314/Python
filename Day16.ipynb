```python
# Q1 Book class attributes title,aurthor,price and method display_info() that print all details

class Book:
    def __init__(self,title,aurthor,price):
        self.title = title
        self.aurthor = aurthor
        self.price = price
    def display_info(self):
        print("Title:",self.title,"Aurthor:",self.aurthor,self.price)

b1 = Book("Meri Kahaani","Ritik",1000)
b1.display_info()
```

```python
#Q4 Rectangle class with length and width , add method area() and perimeter
class Rectangle:
    def __init__(self,l,w):
        self.l = l
        self.w = w
    def area(self):
        print("Area:",self.l*self.w)
    def perimeter(self):
        print("Perimeter:",2*(self.l+self.w))
r1 = Rectangle(10,20)
r1.area()
r1.perimeter()
```

```python
#Q5 create Employee class with name and salary,add method give_raise(percent) 
# that update self.salary in place (no return just modify and print new salary)

class Employee:
    def __init__(self,name,salary):
        self.name = name
        self.salary = salary
    def give_raise(self,per):
        self.salary += self.salary*per/100
    def new_salary(self):
        print(self.salary)

e1 = Employee("Ritik",75000)
e1.give_raise(10)
e1.new_salary()
```

```python
#Q6 create a Library class that holds a list of Books object (from q1).
# Add_book(book) and show_all_books()

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def show_all_books(self):
        for book in self.books:
            book.display_info()

lib = Library()
lib.add_book(Book("Python", "ABC", 500))
lib.add_book(Book("SQL", "XYZ", 700))
lib.show_all_books()
            
```

```python
#incapsulation
#create an employee class with a private attribute __salary. Add a getter and setter -the setter should reject negative salaries.
#Add a method apply_raise(percent) thet increase salary usung the setter inrenally

class Employee:
    def __init__ (self,salary):
        self.__salary = 10000

    @property
    def salary(self):
        return  self.__salary

    @salary.setter
    def salary(self, value):
        if value > 0:
            self.__salary = value
        else:
            print("Salary must be positive.")

    def apply_raise(self,percent):
        self.salary = self.salary + self.salary*percent/100

e1 = Employee(10000)
print(e1.salary)
e1.apply_raise(10)
print(e1.salary)
e1.salary = -100
```
