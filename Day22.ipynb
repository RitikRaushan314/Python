```python
customers = pd.read_csv("C:/Users/raush/ Python Batch/customers.csv")
orders = pd.read_csv("C:/Users/raush/ Python Batch/orders.csv")
sales = pd.read_csv("C:/Users/raush/ Python Batch/sales.csv")
print(customers.head())
print(orders.head())
print(sales.head())
```

```python
jan = orders[orders['OrderDate'] < '2026-02-01']
feb = orders[orders['OrderDate'] >= '2026-02-01']
all_orders = pd.concat([jan, feb], axis=0, ignore_index=True)  
print(all_orders.shape)  
```

```python
print(sales['OrderDate'].dtype)
sales['OrderDate'] = pd.to_datetime(sales['OrderDate'])
print(sales['OrderDate'].dtype)
```

```python
sales['Year'] = sales['OrderDate'].dt.year
sales['Month'] = sales['OrderDate'].dt.month
sales['Day'] = sales['OrderDate'].dt.day
sales['Weekday'] = sales['OrderDate'].dt.day_name()

print(sales[['OrderDate', 'Year', 'Month', 'Day', 'Weekday']].head(3))
```

```python
print(sales[sales['OrderDate'] >= '2026-03-01'][['OrderDate', 'Store', 'Revenue']].head(4))

print(sales[sales['OrderDate'].dt.month == 1][['OrderDate', 'Store', 'Revenue']].head(4))
```

```python
reference_date = pd.Timestamp('2026-07-01')
sales['DaysSinceOrder'] = (reference_date - sales['OrderDate']).dt.days
sales['DueDate'] = sales['OrderDate'] + pd.Timedelta(days=30)
print(sales[['OrderDate', 'DaysSinceOrder', 'DueDate']].head(3))
```
