```python
numbers = (10, 15, 20, 25, 30, 35)

even = 0
odd = 0

i = 0

while i < len(numbers):
    if numbers[i] % 2 == 0:
        even += 1
    else:
        odd += 1
    i += 1

print("Even =", even)
print("Odd =", odd)
```

```python
numbers = (15, 40, 10, 80, 25)

largest = numbers[0]

i = 1

while i < len(numbers):
    if numbers[i] > largest:
        largest = numbers[i]
    i += 1

print("Largest =", largest)
```
