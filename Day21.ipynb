```python
import pandas as pd

# Create DataFrame
data = {
    "Name": ["Ritik", "Raushan", "Aakash", "Ritik", "Aakash"],
    "Age": [20, 21, None, 20, 22],
    "Gender": ["M", "F", "M", "M", "M"],
    "Marks": [85, None, 90, 85, 90]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

# -------------------- Information --------------------
print("\nDataFrame Info")
print(df.info())

print("\nMissing Values")
print(df.isnull())

print("\nMissing Values Count")
print(df.isnull().sum())

print("\nNon-Missing Values")
print(df.notnull())

# -------------------- dropna() --------------------
print("\nDrop Rows with Any Missing Value")
print(df.dropna())

print("\nDrop Rows where Marks is Missing")
print(df.dropna(subset=["Marks"]))

print("\nDrop Rows where All Values are Missing")
print(df.dropna(how="all"))

# -------------------- fillna() --------------------
print("\nFill Missing Marks with 0")
df["Marks"] = df["Marks"].fillna(0)

print("\nFill Missing Age with Mean")
df["Age"] = df["Age"].fillna(df["Age"].mean())

print("\nForward Fill")
df = df.ffill()

print(df)

# -------------------- replace() --------------------
df["Gender"] = df["Gender"].replace({
    "M": "Male",
    "F": "Female"
})

print("\nAfter Replacing Gender")
print(df)

# -------------------- rename() --------------------
df = df.rename(columns={
    "Name": "Student_Name",
    "Age": "Student_Age",
    "Marks": "Student_Marks"
})

print("\nAfter Renaming Columns")
print(df)

# -------------------- duplicated() --------------------
print("\nDuplicate Rows")
print(df.duplicated())

# -------------------- drop_duplicates() --------------------
print("\nRemove Duplicate Rows")
print(df.drop_duplicates())

# -------------------- drop_duplicates(subset=...) --------------------
print("\nRemove Duplicates Based on Student_Name")
print(df.drop_duplicates(subset=["Student_Name"]))

# -------------------- keep='first' --------------------
print("\nKeep First Duplicate")
print(df.drop_duplicates(subset=["Student_Name"], keep="first"))

# -------------------- keep='last' --------------------
print("\nKeep Last Duplicate")
print(df.drop_duplicates(subset=["Student_Name"], keep="last"))

# -------------------- keep=False --------------------
print("\nRemove All Duplicate Rows")
print(df.drop_duplicates(subset=["Student_Name"], keep=False))
```

```python
import pandas as pd

# Create DataFrame
data = {
    "Age": ["20", "21", "22"],            # String
    "Marks": ["85.5", "90.0", "88.5"],    # String
    "Passed": [1, 0, 1],                  # Integer
    "Roll": [101, 102, 103]               # Integer
}

df = pd.DataFrame(data)

print("Original Data Types")
print(df.dtypes)

# -------------------- astype() --------------------

# String -> Integer
df["Age"] = df["Age"].astype(int)

# String -> Float
df["Marks"] = df["Marks"].astype(float)

# Integer -> Boolean
df["Passed"] = df["Passed"].astype(bool)

# Integer -> String
df["Roll"] = df["Roll"].astype(str)

print("\nUpdated Data Types")
print(df.dtypes)

print("\nUpdated DataFrame")
print(df)
```

```python
#sotiing

import pandas as pd

# Create DataFrame
data = {
    "Name": ["Ritik", "Raushan", "Aakash", "Priya", "Neha"],
    "Age": [20, 21, 20, 22, 21],
    "Marks": [85, 90, 85, 95, 88]
}

df = pd.DataFrame(data, index=[4, 2, 0, 3, 1])

print("Original DataFrame")
print(df)

# ---------------- Sort by One Column ----------------
print("\nSort by Age")
print(df.sort_values(by="Age"))

# ---------------- Sort by One Column (Descending) ----------------
print("\nSort by Marks (Descending)")
print(df.sort_values(by="Marks", ascending=False))

# ---------------- Sort by Multiple Columns ----------------
print("\nSort by Age and Marks")
print(df.sort_values(by=["Age", "Marks"]))

# ---------------- Sort by Multiple Columns with Different Order ----------------
print("\nSort by Age (Ascending) and Marks (Descending)")
print(df.sort_values(by=["Age", "Marks"], ascending=[True, False]))

# ---------------- Sort by Index ----------------
print("\nSort by Index")
print(df.sort_index())

# ---------------- Sort by Index (Descending) ----------------
print("\nSort Index (Descending)")
print(df.sort_index(ascending=False))
```

```python
import pandas as pd

# Create DataFrame
data = {
    "Name": ["Ritik", "Raushan", "Aakash"],
    "Age": [20, 21, 22]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

# Add a new column
df["Marks"] = [85, 90, 88]

print("\nAfter Adding Marks Column")
print(df)

# Update an existing column
df["Age"] = [23, 24, 25]

print("\nAfter Updating Age Column")
print(df)

# Add another column
df["Grade"] = ["A", "A+", "B+"]

print("\nAfter Adding Grade Column")
print(df)

# Delete a column
df.drop("Grade", axis=1, inplace=True)

print("\nAfter Deleting Grade Column")
print(df)
```

```python
import pandas as pd

# Create DataFrame
data = {
    "Name": ["Ritik", "Raushan", "Aakash"],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

# ---------------- map() ----------------
# Convert names to uppercase
df["Name"] = df["Name"].map(str.upper)

print("\nAfter map()")
print(df)

# ---------------- apply() ----------------
# Add 5 marks to each student
df["Marks"] = df["Marks"].apply(lambda x: x + 5)

print("\nAfter apply()")
print(df)

# ---------------- applymap() ----------------
# Convert every value in the DataFrame to string
df = df.applymap(str)

print("\nAfter applymap()")
print(df)
```

```python
import pandas as pd

# Create DataFrame
data = {
    "Name": ["Ritik", "Raushan", "Aakash"],
    "Marks": [85, 90, 78],
    "Bonus": [5, 10, 7]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

# Add 5 marks to every student
df["Marks"] = df["Marks"] + 5

print("\nAfter Adding 5 Marks")
print(df)

# Add two columns
df["Total"] = df["Marks"] + df["Bonus"]

print("\nAfter Adding Marks and Bonus")
print(df)

# Multiply Marks by 2
df["Double_Marks"] = df["Marks"] * 2

print("\nAfter Multiplication")
print(df)

# Check students with Marks > 90
df["Result"] = df["Marks"] > 90

print("\nAfter Comparison")
print(df)
```
